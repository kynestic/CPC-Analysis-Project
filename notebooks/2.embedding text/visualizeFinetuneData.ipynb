```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv(r'D:\CODING\Project\NVIDIA-Stock-prediction\data\fine-tuning\keyword_label.csv')

# Count the number of occurrences of each label
label_counts = df['label'].value_counts()

# Limit to top 20 labels (you can adjust the number as needed)
top_labels = label_counts.head(20)

# Optionally, group the remaining labels into 'Other'
if len(label_counts) > 20:
    other_labels = label_counts.iloc[20:].sum()
    top_labels = top_labels.append(pd.Series({'Other': other_labels}))

# Plot the counts using a horizontal bar chart for better readability
plt.figure(figsize=(10, 6))
top_labels.plot(kind='barh', color='skyblue')
plt.title('Top Label Distribution in keyword_label.csv')
plt.xlabel('Count')
plt.ylabel('Label')
plt.show()
```
