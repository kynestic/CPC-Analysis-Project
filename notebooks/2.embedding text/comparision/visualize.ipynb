```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv(r'D:\CODING\Project\NVIDIA Stock prediction\data\embedding_text\dimension reduction\BERT_EmbeddingText_reduction.csv')

x = df.iloc[:, 0]  # First column
y = df.iloc[:, 1]  # Second column

# Create random colors for the scatter plot
N = len(x)  # Number of points to plot
colors = np.random.rand(N)  # Random colors for each point


plt.scatter(x, y, c=colors, alpha=0.5, marker='o')  # 'o' indicates dot marker

# Add labels and title
plt.title('BERT Embeddings Visualization')
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```
