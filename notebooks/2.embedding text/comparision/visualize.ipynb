```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv(r'D:\CODING\Project\NVIDIA Stock prediction\data\embedding_text\dimension reduction\T5_EmbeddingText_reduction.csv')

x = df.iloc[:, 0] 
y = df.iloc[:, 1] 

N = len(x)
colors = np.random.rand(N)


plt.scatter(x, y, c=colors, alpha=0.5, marker='o') 

plt.title('T5 Embeddings Visualization')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
